## match the word

```python
import re
from datetime import datetime, timedelta
def time_helper(time1, time2):
    time1 = time1.split(',')[0]
    time2 = time2.split(',')[0]
    format = '%H:%M:%S'
    time1_obj = datetime.strptime(time1, format)
    time2_obj = datetime.strptime(time2, format)
    diff = (time1_obj - time2_obj).total_seconds()
    if abs(diff) < 5:
        # Expand time span
        new_time1_obj = time1_obj - timedelta(seconds=2)
        new_time2_obj = time2_obj + timedelta(seconds=3)
        time1 = new_time1_obj.strftime(format)
        time2 = new_time2_obj.strftime(format)
        diff = (new_time1_obj - new_time2_obj).total_seconds()
    return time1, time2

def extract_text_info(text):
    # Regular expression pattern to match the text pattern
    pattern = r'(\d+)?\n(\d{2}:\d{2}:\d{2},\d{3}) --> (\d{2}:\d{2}:\d{2},\d{3})\n(.*?)\n\n'
    
    # Find all matches of the pattern in the text
    matches = re.findall(pattern, text, re.DOTALL)
    
    # Create a list of tuples with the extracted information
    
    info_list = [(int(match[0]), *time_helper(match[1],match[2]), match[3]) for match in matches]
    
    return info_list

with open('sub_kor_01.srt','r',encoding='utf-8') as fo:
    res = extract_text_info(fo.read())
```

```python
res = {}
for fn in range(1,17):
    with open(f'sub_kor_{fn:0>2}.srt','r',encoding='utf-8') as fo:
        tmp_res = extract_text_info(fo.read())
        tmp_res = [i for i in tmp_res if  '씨발' in i[3]]
    res[fn] = tmp_res
res
```

## cut the video

```python
cnt=0
for ep,clips in res.items():
    for i in clips:
        cnt+=1
cnt
```

```python
cmds = []
to_fns = []
for ep,clips in res.items():
    cnt = 0
    for i in clips:
        cnt+=1
        to_fn = f'seg_ep_{ep:0>2}_{cnt:0>2}.mp4'
        to_fns.append(to_fn)
        cmds.append(f"ffmpeg -i {ep:0>2}.mp4 -ss {i[1]} -to {i[2]} -c copy {to_fn}")
        
print(*cmds,sep='\n')
    
with open('segments.txt','w') as fo:
    to_fns_text_li = [f"file '{i}'" for i in to_fns]
    fo.write('\n'.join(to_fns_text_li))
```
